```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from functools import reduce
pd.set_option("display.max_rows", None)
```

```python
csv_file = "Resources/COW.csv"
COW_df = pd.read_csv(csv_file)
COW_df.head()
```

```python
csv_file = "Resources/MMR.csv"
MMR_df = pd.read_csv(csv_file)
MMR_df
```

```python
MMR_cleaned = MMR_df.drop(columns = ['Indicator'])
```

```python
MMR_cleaned_df = MMR_cleaned.rename(columns = {'Location':'Country',
                                         'Period':'Year'})
MMR_cleaned_df[['MMR', 'Ranges']] = MMR_cleaned_df['First Tooltip'].str.split(' ', expand=True)
MMR_cleaned_df = MMR_cleaned_df.drop(columns = ['First Tooltip'])
print(MMR_cleaned_df)
```

```python
MMR_country = MMR_cleaned_df['Country'].unique()
MMR_country_df = pd.DataFrame(MMR_country, columns = ['Country'])
print(MMR_country_df)
MMR_country_df.to_csv(r'Resources/MMR_countries.csv', index=False)
```
