```python

import pandas as pd
from sqlalchemy import create_engine
from functools import reduce
pd.set_option("display.max_rows", None)
```

```python
csv_file = "Resources/COW.csv"
COW_df = pd.read_csv(csv_file)
COW_df.head()
```

```python
csv_file = "Resources/MMR.csv"
MMR_df = pd.read_csv(csv_file)
MMR_df.head()
```

```python
COW_cleaned = COW_df.drop(columns = ['Area (sq. mi.)' , 'Pop. Density (per sq. mi.)' , 'Coastline (coast/area ratio)' ,
                                     'Net migration' , 'GDP ($ per capita)' , 'Literacy (%)' , 'Phones (per 1000)' , 'Arable (%)',
                                     'Other (%)', 'Climate', 'Agriculture','Crops (%)','Birthrate','Deathrate','Industry','Service','Infant mortality (per 1000 births)'])
COW_cleaned.head()
```

```python
MMR_cleaned_df = MMR_df.rename(columns = {'Location': 'Country',
                                         'Period':'Year'})                                                      
MMR_cleaned_df[['MMR', 'Ranges']] = MMR_cleaned_df['First Tooltip'].str.split(' ', expand=True)
MMR_cleaned_df = MMR_cleaned_df.drop(columns = ['First Tooltip','Indicator'])
MMR_cleaned_df
```

```python
dfs = [MMR_cleaned_df, COW_cleaned ]

final_df = reduce(lambda  left,right: pd.merge(left,right,on=['Country'],
                                            how='outer'), dfs)

final_df
```

```python
final_df.to_csv(r'Resources/PMD.csv', index = False)

```
